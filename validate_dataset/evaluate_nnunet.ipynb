```python
import os
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt

from medpy.metric import binary
from sklearn.metrics import confusion_matrix, cohen_kappa_score, accuracy_score
from scipy.stats import pearsonr
import pyvista as pv
import seaborn as sns

########################################################################
#                         USER CONFIGURATION                           #
########################################################################

# Ground truth folder remains the same.
GT_DIR = 'nnUNet/val/labelsVal'

# List of prediction folders from different models.
PRED_MODEL_FOLDERS = {
    "Model_001": 'nnUNet/nnUNet_predictions/001',
    "Model_002": 'nnUNet/nnUNet_predictions/002',
    "Model_003": 'nnUNet/nnUNet_predictions/003',
    "Model_004": 'nnUNet/nnUNet_predictions/004',
    "Model_005": 'nnUNet/nnUNet_predictions/005',
    "Model_006": 'nnUNet/nnUNet_predictions/006'
}

# Evaluation will skip the head class (label 1), so we use only these classes.
CLASSES = [2, 3, 4, 5]
CLASS_NAMES = {
    2: "NETC",
    3: "FLAIR",
    4: "EnhancingTumor",
    5: "ResectionCavity",
}

ENABLE_3D_RENDERING = True  # Will render surfaces for one label (see below)

########################################################################
#                        HELPER FUNCTIONS                              #
########################################################################

def dice_coefficient(gt, pred):
    gt = gt.astype(bool)
    pred = pred.astype(bool)
    intersection = np.sum(gt & pred)
    total = np.sum(gt) + np.sum(pred)
    return 1.0 if total == 0 else 2.0 * intersection / total

def jaccard_index(gt, pred):
    gt = gt.astype(bool)
    pred = pred.astype(bool)
    intersection = np.sum(gt & pred)
    union = np.sum(gt | pred)
    return 1.0 if union == 0 else intersection / union

def compute_distance_metrics(gt, pred):
    try:
        hd95 = binary.hd95(pred, gt)
        assd = binary.assd(pred, gt)
    except Exception:
        hd95, assd = np.nan, np.nan
    return hd95, assd

def flatten_for_confusion(gt_data, pred_data):
    assert gt_data.shape == pred_data.shape, "Shape mismatch!"
    return gt_data.flatten(), pred_data.flatten()

def bland_altman_plot(data1, data2, title="Bland-Altman Plot"):
    mean = (data1 + data2) / 2
    diff = data2 - data1
    md = np.mean(diff)
    sd = np.std(diff)
    plt.figure(figsize=(6, 5))
    plt.scatter(mean, diff, c='b', alpha=0.5)
    plt.axhline(md, color='gray', linestyle='--')
    plt.axhline(md + 1.96 * sd, color='red', linestyle='--')
    plt.axhline(md - 1.96 * sd, color='red', linestyle='--')
    plt.title(title)
    plt.xlabel('Mean of Volumes')
    plt.ylabel('Difference (Pred - GT)')
    plt.grid(True)
    plt.show()

########################################################################
#                PROCESSING FOR EACH MODEL                           #
########################################################################

# Get list of ground-truth files
gt_files = sorted([f for f in os.listdir(GT_DIR) if f.endswith('.nii.gz')])

results = {}  # To store aggregated metrics per model

for model_name, pred_dir in PRED_MODEL_FOLDERS.items():
    print(f"\n\n===== Evaluating {model_name} =====\n")
    # Initialize per-model aggregators
    per_class_metrics = {c: [] for c in CLASSES}
    all_conf_gt = []
    all_conf_pred = []
    gt_volumes = []
    pred_volumes = []
    file_count = 0

    for fname in gt_files:
        gt_path = os.path.join(GT_DIR, fname)
        pred_path = os.path.join(pred_dir, fname)
        if not os.path.exists(pred_path):
            print(f"❌ {fname}: Prediction file not found in {pred_dir}, skipping.")
            continue

        gt_img = nib.load(gt_path)
        gt_data = gt_img.get_fdata().astype(np.int32)

        pred_img = nib.load(pred_path)
        pred_data = pred_img.get_fdata().astype(np.int32)

        # Remove head class (label 1) from predictions by mapping it to background (0)
        pred_data[pred_data == 1] = 0

        # For confusion matrix: restrict to evaluation classes (others ignored)
        mask = np.isin(gt_data, CLASSES)
        gt_masked = gt_data[mask]
        pred_masked = pred_data[mask]
        all_conf_gt.extend(gt_masked.flatten())
        all_conf_pred.extend(pred_masked.flatten())

        # Total volume (non-background) for correlation and Bland–Altman.
        gt_nonbg = np.sum(np.isin(gt_data, CLASSES) & (gt_data != 0))
        pred_nonbg = np.sum(np.isin(pred_data, CLASSES) & (pred_data != 0))
        gt_volumes.append(gt_nonbg)
        pred_volumes.append(pred_nonbg)

        # Per-class metrics
        for c in CLASSES:
            gt_bin = (gt_data == c)
            pred_bin = (pred_data == c)
            dsc = dice_coefficient(gt_bin, pred_bin)
            iou = jaccard_index(gt_bin, pred_bin)
            hd95, assd = compute_distance_metrics(gt_bin, pred_bin)
            vol_gt = np.sum(gt_bin)
            vol_pred = np.sum(pred_bin)
            abs_diff = abs(vol_gt - vol_pred)
            rel_diff = abs_diff / vol_gt if vol_gt > 0 else np.nan

            per_class_metrics[c].append({
                'dice': dsc,
                'iou': iou,
                'hd95': hd95,
                'assd': assd,
                'vol_gt': vol_gt,
                'vol_pred': vol_pred,
                'abs_vol_diff': abs_diff,
                'rel_vol_diff': rel_diff
            })

        file_count += 1

    # After processing files for this model, compute confusion matrix and summary metrics.
    all_conf_gt = np.array(all_conf_gt)
    all_conf_pred = np.array(all_conf_pred)
    unique_labels = sorted(list(set(all_conf_gt) | set(all_conf_pred)))
    cm = confusion_matrix(all_conf_gt, all_conf_pred, labels=unique_labels)
    kappa = cohen_kappa_score(all_conf_gt, all_conf_pred, labels=unique_labels)
    overall_accuracy = accuracy_score(all_conf_gt, all_conf_pred)

    print("Confusion Matrix (Rows: GT, Columns: Pred):")
    print(cm)

    tick_labels = [CLASS_NAMES.get(l, str(l)) for l in unique_labels]

    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=tick_labels, yticklabels=tick_labels)
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.title("Confusion Matrix")
    plt.show()
    
    print(f"Cohen's Kappa: {kappa:.4f}")
    print(f"Overall Accuracy: {overall_accuracy:.4f}")

    # Per-class confusion-based metrics
    eps = 1e-8
    for idx, label in enumerate(unique_labels):
        tp = cm[idx, idx]
        fn = np.sum(cm[idx, :]) - tp
        fp = np.sum(cm[:, idx]) - tp
        tn = np.sum(cm) - (tp + fp + fn)
        precision = tp / (tp + fp + eps)
        recall = tp / (tp + fn + eps)
        specificity = tn / (tn + fp + eps)
        f1 = 2 * precision * recall / (precision + recall + eps)
        print(f"\nClass {label} ({CLASS_NAMES.get(label, 'Ignored')}):")
        print(f"  Precision:  {precision:.4f}")
        print(f"  Recall:     {recall:.4f}")
        print(f"  Specificity:{specificity:.4f}")
        print(f"  F1-score:   {f1:.4f}")

    # Average per-class overlap and distance metrics
    print("\nAverage per-class metrics:")
    for c in CLASSES:
        metrics_list = per_class_metrics[c]
        dice_vals = [m['dice'] for m in metrics_list]
        iou_vals = [m['iou'] for m in metrics_list]
        hd95_vals = [m['hd95'] for m in metrics_list if not np.isnan(m['hd95'])]
        assd_vals = [m['assd'] for m in metrics_list if not np.isnan(m['assd'])]
        mean_dice = np.mean(dice_vals)
        mean_iou = np.mean(iou_vals)
        mean_hd95 = np.mean(hd95_vals) if hd95_vals else np.nan
        mean_assd = np.mean(assd_vals) if assd_vals else np.nan
        print(f"  Class {c} ({CLASS_NAMES.get(c)}): Dice={mean_dice:.4f}, IoU={mean_iou:.4f}, HD95={mean_hd95:.4f}, ASSD={mean_assd:.4f}")

    # Volume correlation and Bland–Altman
    gt_volumes = np.array(gt_volumes, dtype=float)
    pred_volumes = np.array(pred_volumes, dtype=float)
    corr_coeff, p_val = pearsonr(gt_volumes, pred_volumes)
    print("\nVolume Comparison:")
    print(f"  Pearson's r = {corr_coeff:.4f} (p={p_val:.4e})")

    plt.figure(figsize=(6, 5))
    plt.scatter(gt_volumes, pred_volumes, alpha=0.6)
    plt.title(f"{model_name}: Scatter of Total Volumes (voxels)")
    plt.xlabel("GT Volume")
    plt.ylabel("Pred Volume")
    plt.grid(True)
    plt.show()

    bland_altman_plot(gt_volumes, pred_volumes, title=f"{model_name}: Bland–Altman Plot")

    results[model_name] = {
        "confusion_matrix": cm,
        "kappa": kappa,
        "accuracy": overall_accuracy,
        "per_class": per_class_metrics,
        "volume_correlation": (corr_coeff, p_val)
    }

########################################################################
#                3D SURFACE RENDERING EXAMPLE                          #
########################################################################

# For demonstration, use the first file from GT and the predictions from one model.
example_file = gt_files[0] if gt_files else None
if ENABLE_3D_RENDERING and example_file:
    # Choose one model to render (e.g., Model_001)
    chosen_model = list(PRED_MODEL_FOLDERS.keys())[0]
    pred_path = os.path.join(PRED_MODEL_FOLDERS[chosen_model], example_file)
    gt_path = os.path.join(GT_DIR, example_file)
    gt_data = nib.load(gt_path).get_fdata().astype(np.int32)
    pred_data = nib.load(pred_path).get_fdata().astype(np.int32)
    pred_data[pred_data == 1] = 0  # Remove head class

    # Render 3D surfaces for one label (e.g., label=3: EnhancingTumor)
    label_to_render = 3
    gt_bin = (gt_data == label_to_render)
    pred_bin = (pred_data == label_to_render)

    # Create PyVista meshes using marching cubes (assumes isotropic voxels)
    gt_volume = pv.wrap(gt_bin.astype(np.uint8))
    pred_volume = pv.wrap(pred_bin.astype(np.uint8))
    try:
        gt_mesh = gt_volume.contour(isosurfaces=[0.5])
        pred_mesh = pred_volume.contour(isosurfaces=[0.5])
    except Exception as e:
        print("3D rendering failed:", e)
        gt_mesh = pred_mesh = None

    if gt_mesh is not None and pred_mesh is not None:
        p = pv.Plotter()
        p.add_mesh(gt_mesh, color='blue', opacity=0.5, label='GT')
        p.add_mesh(pred_mesh, color='red', opacity=0.5, label='Pred')
        p.add_legend()
        p.show(title=f"3D Surface Rendering (Label {label_to_render})")

print("\nAll computations complete.")
```
